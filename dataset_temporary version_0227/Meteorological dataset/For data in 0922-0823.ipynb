Solar Irradiance data

```python
import os
import pandas as pd

# Specify the folder path using a raw string literal
folder_path = r'C:\Users\nickc\OneDrive - HKUST Connect\桌面\FYP Code\dataset_temporary version_0227\Meteorological dataset\Irradiance'

# Get all the CSV files in the folder
excel_files = [file for file in os.listdir(folder_path) if file.endswith('.xlsx')]

# Read and display the data from each excel file
for file in excel_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_excel(file_path)
    print(f"Data from {file}:")
    print(df)
    print()
print(df.columns)
```

```python
# Combine all the data into a single DataFrame
combined_df = pd.concat([pd.read_excel(os.path.join(folder_path, file)) for file in excel_files], ignore_index=True)
combined_df 
```

```python
start_date = '2022-09-01 00:00'
end_date = '2023-08-31 23:59'

filtered_df = combined_df.loc[combined_df['Time'].between(start_date, end_date)]
filtered_df
```

```python
# Resample the dataframe to 15-minute intervals and calculate the mean
mean_df = filtered_df.resample('15T', on='Time').mean()

# Display the mean dataframe
mean_df
```

```python
mean_df.to_excel('required irradiance.xlsx')
```
