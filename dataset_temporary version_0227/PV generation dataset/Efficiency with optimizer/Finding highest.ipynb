```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
#read an csv file
df = pd.read_excel('All Efficiency.xlsx')
df 
```

```python
#I only want data from 7am to 19am of each day
#new_df = df[(df['Time'] >= '2022-09-01 07:00:00') & (df['Time'] <= '2023-08-31 19:00:00')]
#new_df
start_date = '2022-09-01 00:00:00'
end_date = '2023-08-31 23:45:00'
filtered_df = df.loc[df['Time'].between(start_date, end_date)]
filtered_df

#Only include data from 7am to 7pm of each day
filtered_df['Time'] = pd.to_datetime(filtered_df['Time'])
filtered_df = filtered_df.set_index('Time')
filtered_df = filtered_df.between_time('07:00:00', '19:00:0')
filtered_df = filtered_df.reset_index()
filtered_df
filtered_df.to_excel('All Efficiency @7am-7pm.xlsx', index=False)
```
