```python
import pandas as pd
import os
import datetime as dt
import numpy as np
```

```python
#read an csv file
df = pd.read_csv(r'PV generation dataset\output without optimizer with area\SQ11.csv')
df['Efficiency'] = df['power(W)']*100/(df['Irradiance (W/m2)']*df['Area'])
df
```

```python
#read all csv files in a folder
path = r'PV generation dataset\Testing'
files = os.listdir(path)
for file in files:
    df = pd.read_csv(path + '\\' + file)
    df['Efficiency'] = df['power(W)']*100/(df['Irradiance (W/m2)']*df['Area'])
    df.to_csv(path + '\\' + file, index = False)
    print(file + ' is done')
```
