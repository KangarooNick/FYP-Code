```python
#url = "https://hkust.azure-api.net/energy/generation/_search?q=(location:'UST Library'%20AND%20@timestamp:[2023-09-26%20TO%202023-10-26])&size=10000"
url = "https://hkust.azure-api.net/energy/generation/_search?q=(@timestamp:[2023-11-20%20TO%202023-11-26])&size=10000"
#url = "https://hkust.azure-api.net/energy/generation/_search?q=(location:'Chia_Wei_Woo_Concourse_Academic_Zone_J2'%20AND%20@timestamp:[2023-09-26%20TO%202023-09-26])&size=10000"
payload={}
headers = {
  'X-Apim-Subscription-Key': '36dc61a01022434a9e1038668665d8ed'
}

response = requests.request("GET", url, headers=headers, data=payload)

json_data = response.json()['hits']['hits']

pprint.pprint(json_data)
```

```python
data_set = []

for i in json_data:
    data_dict = {}
    if 'type' in i['_source']:
        data_dict['type'] = i['_source']['type']
    else: data_dict['type'] = " "
    if 'location' in i['_source']:
        data_dict['location'] = i['_source']['location']
    else: data_dict['location'] = " "
    if 'generation' in i['_source']:
        data_dict['generation'] = i['_source']['generation']
    else: data_dict['generation'] = " "
    if '@timestamp' in i['_source']:
        data_dict['@timestamp'] = i['_source']['@timestamp']
    else: data_dict['@timestamp'] = " "
# You can add any other information you need with another if statement

    data_set.append(data_dict)
    
df = pd.DataFrame(data_set)

df
```
