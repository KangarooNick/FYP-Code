```python
import pandas as pd
import matplotlib.pyplot as plt
```

Temperature and Energy Generation

```python
data1 = pd.read_excel('Temperature Monthly.xlsx')
```

```python
print(data1)
```

```python
data1.plot(x='Month', y='Arithmetic mean', kind='line')
plt.ylim(0, 30)
plt.xlabel('Month')
plt.ylabel('Average temperature')
plt.show()
```

```python
data2 = pd.read_excel('Monthly Mean Energy Generation.xlsx')
```

```python
print(data2)
```

```python
data2.plot(x='Month', y='Mean', kind='line')
plt.xlabel('Month')
plt.ylabel('Average Energy generation')
plt.show()
```

```python
# Create the figure and the first y-axis
fig, ax1 = plt.subplots()

# Plot the first dataset on the first y-axis
ax1.plot(data1['Month'], data1['Arithmetic mean'], color='blue')
ax1.set_xlabel('Month')
ax1.set_ylabel('Average Temperature', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create the second y-axis
ax2 = ax1.twinx()

# Plot the second dataset on the second y-axis
ax2.plot(data2['Month'], data2['Mean'], color='red')
ax2.set_ylabel('Average Energy Generation', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Customize the graph
plt.title('Relationship between Temperature and Energy Generation')
plt.legend()

# Show the plot
plt.show()
```

```python
# Calculate the correlation coefficient
correlation = data1['Arithmetic mean'].corr(data2['Mean'])

print("Correlation coefficient:", correlation)
```
