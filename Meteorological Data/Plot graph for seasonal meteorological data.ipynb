```python
import pandas as pd
import matplotlib.pyplot as plt
```

Temperature and Energy Generation

```python
data1 = pd.read_excel('Temperature Monthly.xlsx')
```

```python
print(data1)
```

```python
data1.plot(x='Month', y='Arithmetic mean', kind='line')
plt.ylim(0, 30)
plt.xlabel('Month')
plt.ylabel('Average temperature')
#add points on the data
plt.show()
```

```python
#plot a bar chart
data1.plot(x='Month', y='Arithmetic mean', kind='bar')
plt.ylim(0, 30)
plt.xlabel('Month')
plt.ylabel('Average temperature')
plt.show()
```

```python
data5 = pd.read_excel('Monthly Mean Energy Generation.xlsx')
```

```python
print(data5)
```

```python
data5.plot(x='Month', y='Mean', kind='line')
plt.xlabel('Month')
plt.ylabel('Average Energy generation')
plt.show()
```

```python
#plot a bar chart
data5.plot(x='Month', y='Mean', kind='bar')
plt.xlabel('Month')
plt.ylabel('Average Energy generation')
plt.show()
```

```python
# Create the figure and the first y-axis
fig, ax1 = plt.subplots()

# Plot the first dataset on the first y-axis
ax1.plot(data1['Month'], data1['Arithmetic mean'], color='blue')
ax1.set_xlabel('Month')
ax1.set_ylabel('Average Temperature (Degree Celsius)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Add the blue dots on the line
ax1.scatter(data1['Month'], data1['Arithmetic mean'], color='blue')

# Create the second y-axis
ax2 = ax1.twinx()

# Plot the second dataset on the second y-axis
ax2.plot(data5['Month'], data5['Mean'], color='red')
ax2.set_ylabel('Average Energy Generation (kWh)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Customize the graph
plt.title('Relationship between Temperature and Energy Generation')
plt.legend()

#add the points on the data
ax1.scatter(data1['Month'], data1['Arithmetic mean'], color='blue')
plt.scatter(data5['Month'], data5['Mean'], color='red')

# Show the plot
plt.show()
```

```python
# Calculate the correlation coefficient
correlation = data1['Arithmetic mean'].corr(data5['Mean'])

print("Correlation coefficient:", correlation)
```

Relative Humidity and Energy Generation

```python
data3 = pd.read_excel('RH monthly.xlsx')
print(data3)
```

```python
data3.plot(x='Month', y='Arithmetic Mean', kind='line')
plt.ylim(30,100)
plt.xlabel('Month')
plt.ylabel('Average Relative Humidity')
plt.show()
```

```python
# Create the figure and the first y-axis
fig, ax1 = plt.subplots()

# Plot the first dataset on the first y-axis
ax1.plot(data3['Month'], data3['Arithmetic Mean'], color='blue')
ax1.set_xlabel('Month')
ax1.set_ylabel('Average Relative Humidity(%)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create the second y-axis
ax2 = ax1.twinx()

# Plot the second dataset on the second y-axis
ax2.plot(data5['Month'], data5['Mean'], color='red')
ax2.set_ylabel('Average Energy Generation(kWh)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

#add the points on the data
ax1.scatter(data3['Month'], data3['Arithmetic Mean'], color='blue')
plt.scatter(data5['Month'], data5['Mean'], color='red')

# Customize the graph
plt.title('Relationship between Relative Humidity and Energy Generation')
plt.legend()
plt.show()
```

```python
# Calculate the correlation coefficient
correlation = data3['Arithmetic Mean'].corr(data5['Mean'])

print("Correlation coefficient:", correlation)
```

Solar Irradiance and Energy Generation

```python
data4 = pd.read_excel('Daily Solar irradiance.xlsx')
print(data4)
```

```python
monthly_mean = data4.groupby(data4['Time'].dt.month)['W/m2'].mean()
monthly_mean
```

```python
monthly_mean.to_excel('Monthly Irradiance.xlsx')
```

```python
data4 = pd.read_excel('Monthly Irradiance.xlsx')
print(data4)
```

```python
# Create the figure and the first y-axis
fig, ax1 = plt.subplots()

# Plot the first dataset on the first y-axis
ax1.plot(data4['Time'], data4['W/m2'], color='blue')
ax1.set_xlabel('Time')
ax1.set_ylabel('Average Solar Irradiance(W/m2)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create the second y-axis
ax2 = ax1.twinx()

# Plot the second dataset on the second y-axis
ax2.plot(data5['Month'], data5['Mean'], color='red')
ax2.set_ylabel('Average Energy Generation(kWh)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

#add the points on the data
ax1.scatter(data4['Time'], data4['W/m2'], color='blue')
plt.scatter(data5['Month'], data5['Mean'], color='red')

# Customize the graph
plt.title('Relationship between Solar Irradiance and Energy Generation')
plt.legend()
plt.show()
```
