```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#read excel files
Energy_generation= pd.read_excel('Daily generation.xlsx')
Temperature= pd.read_excel('Average Temperature-Daily.xlsx')
Relative_Humidity= pd.read_excel('Average Relative Humidity Daily.xlsx')
Irradiance= pd.read_excel('Daily Solar irradiance.xlsx')
Energy_generation
```

Temperature VS Energy Generation

```python
# Plot a scatter plot graph with temperature as x-axis and energy generation as y-axis
plt.scatter(Temperature['Degree Celsius'], Energy_generation['generation'])
plt.xlabel('Temperature(Degree Celsius)')
plt.ylabel('Energy Generation(kWh)')
plt.title('Energy Generation vs Temperature')
plt.show()
```

Relative Humidity VS Energy Generation


```python
# Plot a scatter plot graph with relative humidity as x-axis and energy generation as y-axis
plt.scatter(Relative_Humidity['%'], Energy_generation['generation'])
plt.xlabel('Relative Humidity(%)')
plt.ylabel('Energy Generation(kWh)')
plt.title('Energy Generation vs Relative Humidity')
plt.show()
```

```python
# Plot a scatter plot graph with irradiance as x-axis and energy generation as y-axis
plt.scatter(Irradiance['W/m2'], Energy_generation['generation'])
plt.xlabel('Irradiance (W/m2)')
plt.ylabel('Energy Generation(kWh)')
plt.title('Energy Generation vs Irradiance')
plt.show()
```
