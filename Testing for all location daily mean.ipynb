```python
import pandas as pd
import os

# Specify the folder path
folder_path = 'Whole Year Testing'

# Create an empty list to store the modified DataFrames
modified_dataframes = []

# Read and process each Excel file in the folder
for file in os.listdir(folder_path):
    if file.endswith('.xlsx'):
        file_path = os.path.join(folder_path, file)
        df = pd.read_excel(file_path, parse_dates=['@timestamp'], index_col='@timestamp')
        
        # Delete the two unwanted columns
        columns_to_delete = ['type', 'location']
        df = df.drop(columns_to_delete, axis=1)
        
        # Remove rows with a data value of "0"
        df = df[df != 0].dropna()
        
        # Calculate the mean at daily resolution
        daily_mean = df.resample('D').mean()
        
        # Append the modified DataFrame to the list
        modified_dataframes.append(daily_mean)

# Concatenate the list of modified DataFrames into a single DataFrame
combined_df = pd.concat(modified_dataframes)

# Calculate the mean for each day separately
daily_means = combined_df.mean(axis=1)

# Print the daily mean values
print("Daily Mean Values:")
print(daily_means)
```

```python
import pandas as pd
import os

# Specify the folder path
folder_path = 'Whole Year Testing'

# Create an empty DataFrame to store the combined data
combined_df = pd.DataFrame()

# Read and process each Excel file in the folder
for file in os.listdir(folder_path):
    if file.endswith('.xlsx'):
        file_path = os.path.join(folder_path, file)
        df = pd.read_excel(file_path, parse_dates=['@timestamp'])
        df = df.set_index('@timestamp')
        
        # Delete the two unwanted columns
        columns_to_delete = ['type', 'location']
        df = df.drop(columns_to_delete, axis=1)
        
        # Remove rows with a data value of "0"
        df = df[df != 0].dropna()
        
        # Merge the current DataFrame with the combined DataFrame
        if combined_df.empty:
            combined_df = df
        else:
            combined_df = combined_df.merge(df, how='outer', left_index=True, right_index=True, suffixes=('', '_{}'.format(file)))

# Calculate the mean between the columns
combined_df['Mean'] = combined_df.mean(axis=1)

# Print the combined DataFrame
print(combined_df)
df.to_excel('Testing.xlsx', index=True)
```
