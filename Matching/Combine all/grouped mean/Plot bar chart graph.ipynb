```python
import pandas as pd
import matplotlib.pyplot as plt
```

Rainfall(mm)

```python
# Load the data from the Excel file
data_rainfall = pd.read_excel('grouped_mean_generation_rainfall.xlsx')
data_rainfall.plot(x='Rainfall(mm)', y='generation', kind='bar')
#add unit for the axis
plt.xlabel('Rainfall(mm)')
plt.ylabel('Energy Generation(kWh)')
plt.show()
```

Temperature

```python
# Load the data from the Excel file
data_temperature = pd.read_excel('grouped_mean_generation_temperature.xlsx')
data_temperature.plot(x='temperature', y='generation', kind='bar')
plt.xlabel('Temperature (C)')
plt.ylabel('Energy Generation(kWh)')
plt.show()
```

Relative humidity

```python
# Load the data from the Excel file
data_temperature = pd.read_excel('grouped_mean_generation_relative humidity.xlsx')
data_temperature.plot(x='relative humidity', y='generation', kind='bar')
plt.xlabel('Relative Humidity(%)')
plt.ylabel('Energy Generation(kWh)')
plt.show()
```

Solar Irradiance

```python
# Load the data from the Excel file
data_temperature = pd.read_excel('grouped_mean_generation_solar irradiance.xlsx')
data_temperature.plot(x='W/m2', y='generation', kind='bar')
plt.xlabel('Irriadiance(W/m2)')
plt.ylabel('Energy Generation(kWh)')
plt.show()
```

Temperature VS Solar Irradiance

```python
# Load the data from the Excel file
data_temperature = pd.read_excel('grouped_mean_solar irradiance_temperature.xlsx')
data_temperature.plot(x='temperature', y='W/m2', kind='bar')
plt.xlabel('Temperature(C)')
plt.ylabel('Irradiance(W/m2)')
plt.show()
```

```python
#plot the scatter plot graph
data = pd.read_excel('grouped_mean_.xlsx')
data.plot.scatter(x='Rainfall(mm)', y='generation')
plt.xlabel('Rainfall(mm)')
plt.ylabel('Energy Generation(kWh)')
plt.show()
```

```python
data= pd.read_excel('Combine all.xlsx')
correlation_factor_1 = data['temperature'].corr(data['W/m2'])
correlation_factor_2 = data['generation'].corr(data['W/m2'])
correlation_factor_3 = data['generation'].corr(data['temperature'])
print("correlation factor between temperature and solar irradiance=", correlation_factor_1)
print("correlation factor between energy generation and solar irradiance=", correlation_factor_2)
print("correlation factor between energy generation and temperature=", correlation_factor_3)
```
