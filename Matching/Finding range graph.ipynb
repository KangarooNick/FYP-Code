```python
import pandas as pd

# Read the Excel file
df = pd.read_excel('Combine all.xlsx')

# Filter the data based on temperature ranges
temperature_ranges = [(10, 15), (15, 20), (20, 25), (25, 30)]
result_df = pd.DataFrame(columns=['Temperature Range', 'Generation Data'])

# Iterate over temperature ranges
for temp_range in temperature_ranges:
    min_temp, max_temp = temp_range
    
    # Filter the data based on the temperature range
    filtered_df = df[(df['temperature'] >= min_temp) & (df['temperature'] <= max_temp)]
    
    # Retrieve the corresponding generation data
    generation_data = filtered_df['generation']
    
    # Create a temporary dataframe for the current temperature range
    temp_df = pd.DataFrame({
        'Temperature Range': [temp_range],
        'Generation Data': [str(generation_data.tolist())]
    })
    # Concatenate the temporary dataframe with the result dataframe
    result_df = pd.concat([result_df, temp_df], ignore_index=True)

# Create a new dataframe with each data value in a separate cell
final_df = pd.DataFrame(columns=['Temperature Range', 'Generation Data'])
for idx, row in result_df.iterrows():
    temp_range = row['Temperature Range']
    generation_data = row['Generation Data']
    
    for data in generation_data:
        final_df = pd.concat([final_df, pd.DataFrame({'Temperature Range': [temp_range], 'Generation Data': [data]})], ignore_index=True)

# Export the final dataframe to an Excel file
final_df.to_excel('Testing.xlsx', index=False)
```
